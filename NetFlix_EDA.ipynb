# Lets do exploratory data analysis for the dataset Netflix 

Installing and inputing the modules that the data will need for the use

Dataset reading and decribtion doing needs numpy, pandas and matplotlib

```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('netflix_titles.csv')
data.head()
```

```python
data.info()
```

```python
data.describe(include='object')
```

```python
data.columns
```

```python
data.isna().sum()
```

```python
data['director'].replace(np.nan,'No Director',inplace=True)
data['cast'].replace(np.nan,'No Cast',inplace=True)
data['country'].replace(np.nan,'No country',inplace=True)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data['rating'].value_counts()
```

```python
data['rating'].unique()
```

```python
data['rating'] = data['rating'].replace({'74 min': np.nan, '84 min':np.nan, '66 min':np.nan, 'G':'TV-G'})
```

```python
data['rating'].unique()
```

```python
data['type'].value_counts()
```

```python
listed_in = data['listed_in'].value_counts()
print(listed_in.to_string())
```

```python
#fill null values with mode
mode_im = ['date_added','rating','duration']
for i in mode_im:
    data[i] = data[i].fillna(data[i].mode()[0])
```

```python
#drop useless columns
data.drop(['show_id','date_added','description'],axis='columns', inplace= True)
data.isnull().sum()
```

# Data Visualization

```python
data.head()
```

```python
plt.figure(figsize=(5,5))
colors =['tab:green','tab:blue']
sns.countplot(data=data,x = 'type',palette=colors)
plt.title("No Of Movies vs TV Shows")
plt.show()
```

```python
plt.figure(figsize=(5,5))
colors =['tab:green','tab:blue']
sns.countplot(data=data,x = 'rating',palette=colors)
plt.title("Released with highest ranking")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=data,y='country', order=data['country'].value_counts().index[:15])
plt.tick_params(labelsize=10)
plt.title('Country with most released')
plt.show()
```
